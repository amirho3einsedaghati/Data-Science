# method chaining

The first of pandorable is called **method chaining**.

The general idea behind method chaining is that **every method on an object returns a reference to that object**. The beauty of this is that **we can condense many different operations on a DataFrame**, for instance, into one line or at least one statement of code.

**In the method chaining, there's no in place flag being used**, and we can see that when we first run a where query, then a dropna, then a set_index, and then a rename. we might wonder why **the whole statements are enclosed in parentheses** it makes the statement more readable.

```python
(df.where(df["SUMLEV"] == 50)
    .dropna()
    .set_index(["STNAME", "CTYNAME"])
    .rename(columns= {"ESTIMATESBASE2010" : "estimates base 2010"}))
```

this code is a traditional way of writing top code that don't have any problem in the fucntional way but it's not a pandorable.

```python
new_df = df[df["SUMLEV"] == 50]
new_df.set_index(["STNAME", "CTYNAME"], inplace= True)
new_df = new_df.rename(columns= {"ESTIMATESBASE2010" : "estimate base 2010"})
new_df
```

which one runs faster, and when do we use chaining method ?

We can put the approach into a function and **pass the function** into the **.timeit()** function **to count the time** the **parameter number allows us to choose how many times we want to run the function**.

```python
def first_approach():
    
    return (df.where(df["SUMLEV"] == 50)
                .dropna()
                .set_index(["STNAME", "CTYNAME"])
                .rename({"ESTIMATESBASE2010" : "estimate base 2010"}))

timeit.timeit(first_approach, number=1)
```

```python
def second_approach():

    new_df = df[df["SUMLEV"] == 50]
    new_df = new_df.set_index(["STNAME", "CTYNAME"])
    new_df = new_df.rename({"ESTIMATESBASE2010" : "estimate base 2010"})
    return new_df

timeit.timeit(second_approach, number=1)
```

As you can see, the second approach is much faster!

# Apply

**Python** has a wonderful function called **map**, which is sort of a basis for functional programming in the language. When we want to use map in Python, **we pass it some function you want called, and some iterable**, like a list, **that you want the given function to be applied to**. **the given function is called against each item in the iterable**. The results will return by a map object including a list of all of the evaluations of that function.

**Python** has a similar function like map called applymap. **In applymap, we provide some function which should operate on each cell of a DataFrame**, and **the return set is itself a DataFrame**.

The second of pandorable is called **.apply() method**.

**the pandas DataFrame and Series objects** has a method called **.apply()**. **this method takes a fucntion to call it against each row in the data frame**, and takes the **axis** on which to operate as parameters.

**to apply across all rows in DataFrame object**, which is applying on all columns, **we pass axis equal to one**.

In cencus DataFrame, we have five columns for population estimates. Each column corresponding with one year of estimates. It's quite reasonable to want to create some new columns for minimum or maximum values, and the **apply** 
function is an easy way to do this.

by using the below code, we can return a Series object with max and min to display it by a new DataFrame object.

```python
def min_max(row):
    new_df = row[["POPESTIMATE2010",
                 "POPESTIMATE2011",
                 "POPESTIMATE2012",
                 "POPESTIMATE2013",
                 "POPESTIMATE2014",
                 "POPESTIMATE2015"]]
    
    return pd.Series({"min" : np.min(new_df), "max" : np.max(new_df)})
```

```python
df.apply(min_max, axis= 1)
```

if we're doing as a part of data cleaning, it's better to add new data to the existing DataFrame.

Instead of returning a separate series to display the min and max using a new DataFrame object. We can add two new columns in the original dataframe to store min and max.

```python
def min_max(row):
    new_df = row[["POPESTIMATE2010",
                 "POPESTIMATE2011",
                 "POPESTIMATE2012",
                 "POPESTIMATE2013",
                 "POPESTIMATE2014",
                 "POPESTIMATE2015"]]
    
    row["min"] = np.min(new_df)
    row["max"] = np.max(new_df)
    
    return row

    raise ImportError()
```

```python
df.apply(min_max, axis= "columns")
```

we rarely see **.apply()** used with **large function** definitions, like we did. Instead, we typically see it used with **lambdas**.

we can imagine **how we might chain several apply calls with lambdas together** to create a readable yet succinct data manipulation script.

the result show us by a Series object because we only have max values.

```python
columns = ["POPESTIMATE2010",
          "POPESTIMATE2011",
          "POPESTIMATE2012",
          "POPESTIMATE2013",
          "POPESTIMATE2014",
          "POPESTIMATE2015"]

df.apply(lambda x: np.max(x[columns]), axis= 1)
```

The beauty of the apply function is that it allows flexibility in doing **whatever manipulation that we desire**, and **the function we pass into apply can be any customized function that you write**. Let's say we want to divide the states into four categories: Northeast, Midwest, South, and West.

We can write a customized function that returns the region based on the state regions information is obtained from Wikipedia.

```python
def get_state_region(state):
    
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    
    if state in northeast:
        return "Northeast"
    elif state in midwest:
        return "Midwest"
    elif state in south:
        return "South"
    else:
        return "West"
```

we can create a new column called state region, which shows the state's region, by using the customized function and the apply function.

```python
df["state region"] = df["STATE"].apply(lambda x : get_state_region(x))
df["state region"]
```
