We can actually use the function .isnull() to create a boolean mask of the whole dataframe. This effectively broadcasts the isnull() function to every cell of data.

```python
# we create an Boolean Mask with .isnull()
mask = df.isnull()
mask.head(10)
```

to drop all of those rows which have any missing data, we should use .dropna() .

```python
df.dropna().head(10)
```

One of the handy functions that Pandas has for working with missing values is the filling function, fillna(). This function takes a number or parameters. You could pass in a single value which is called a scalar value to change all of the missing data to one value.

```python
df.fillna(0, inplace= True)
df.head(10)
```

```python
df = pd.read_csv("datasets/log.csv")
df
```

in the .fillna(), there is a parameter called **method** and we can fill it with 2 values "ffill and bfill".

ffill is for forward filling and it updates an na value for a particular cell with the value from the previous row. bfill is
backward filling, which is the opposite of ffill. It fills the missing values with the next valid value.

It's important to note that your data needs to be sorted in order for this to have the effect you might want. so before using ffill or bfill, we should sort our data. 

```python
df.set_index(keys= "time", inplace= True)
df.sort_index(inplace= True)
df.head(10)
```

If we look closely at the output though we'll notice that the index isn't really unique. Two users seem to be able to use the system at the same time. Again, a very common case. Let's reset the index, and use some multi-level indexing on time and user together instead, promote the user name to a second level of the index to deal with that issue.

```python
df.reset_index(inplace= True)
df.set_index(["time", "user"], inplace= True)
df.head(20)
```

```python
df = df.fillna(method= "ffill")
df.head(10)
```

# exercise :

imagine we want to detect all html pages in the "video" column, lets say that just means they end with ".html", and we want to overwrite that with the keyword "webpage". How could we accomplish this?

```python
df = pd.read_csv("datasets/log.csv")
df
```

To replace using a regex we make the first parameter to replace the regex pattern we want to match, the second parameter the value we want to emit upon match, and then we pass in a third parameter "regex=True".

<br>
solution 1:

```python
df.replace(".+\.\w+", "webpage", regex= True).head(10)
```

<br>
solution 2 :


```python
df.replace(to_replace= ".*.html$", value= "webpage", regex= True).head(10)
```
