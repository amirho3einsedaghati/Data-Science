```python
matrix = np.array([[1.5, 2, 4], [12, 0.5, 1]])
matrix
```

```python
matrix[1,1]
```

```python
print(np.array([matrix[0,0], matrix[1,2], matrix[1,1]]))

print(matrix[[0,1,1], [0,2,1]])
```

# 3.2. Boolean Indexing

```python
matrix > 5
```

```python
print(matrix[matrix > 5])

print(vector[vector > 5])
```

# 3.3. Array Slicing

```python
vector = np.array([1,2,4,5,10,12])
print(vector)
print(vector[2])
```

```python
matrix = np.array([[1, 0.75, 12], [7, 17, 2], [4, 0.125, 5]])
print(matrix)
```

```python
print(matrix[:2])
```

```python
print(matrix[:2, 1:3])
```

```python
sub_array = matrix[:2, 1:3]
print("sub-array index [0, 0] value before change =", sub_array[0, 0])
sub_array[0, 0] = 100
print("sub-array index [0, 0] value before change =", sub_array[0, 0])
print("original-array index [0 , 1] value before change =", matrix[0 ,1])

```

### 4.1. to get the data of a certain column or row or a specific item of an array using array slicing

<br>
access to a certain column

<br>
if we want to have several values of an array, the below command will transform a 2-demensional array to a single-demensional array and then will return it.

```python
print(wines[:, 0])
```

<br>
if we want to have several values of an array, the below command will return a 2-demensional array.

```python
print(wines[:, 0:1])
```

<br>
access to a certain row

<br>
if we want to have several values of an array, the below command will transform a 2-demensional array to a single-demensional array and then will return it.

```python
print(wines[0, :])
```

<br>
if we want to have several values of an array, the below command will return a 2-demensional array.

```python
print(wines[0:1, :])
```

access to a specefic item

<br>
because we want to access one of the values of the array, the below command will return a single-demensional array of it.

```python
print(wines[0:1, 0])
```

<br>
because we want to access one of the array values, the below command will return a single-demensional array of it.

```python
print(wines[0, 0])
```
